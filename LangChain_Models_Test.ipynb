# this code connects to the "HuggingFaceH4/zephyr-7b-beta" model hosted on the Hugging Face Inference API and uses it to answer the question "What is the capital of India" through LangChain's chat model interface.

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv

load_dotenv()

llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation"
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("What is the capital of India")

print(result.content)
```

#  this code downloads the "TinyLlama/TinyLlama-1.1B-Chat-v1.0" model, sets up a local text generation pipeline with it, and then uses this pipeline through LangChain's ChatHuggingFace wrapper to answer a question.

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
import os

os.environ['HF_HOME'] = '/content/sample_data/huggingface_cache'

llm = HuggingFacePipeline.from_model_id(
    model_id='TinyLlama/TinyLlama-1.1B-Chat-v1.0',
    task='text-generation',
    pipeline_kwargs=dict(
        temperature=0.5,
        max_new_tokens=100
    )
)
model = ChatHuggingFace(llm=llm)

result = model.invoke("What is the capital of India")

print(result.content)
```

# Documents similarity Score

**this code takes a query and a set of documents, converts them into numerical representations (embeddings), calculates how similar the query embedding is to each document embedding, and then identifies and prints the document that is most similar to the query based on cosine similarity.**

```python
# Imports the class for using OpenAI's embedding models with LangChain.
from langchain_openai import OpenAIEmbeddings
# Imports the function to load environment variables (like API keys).
from dotenv import load_dotenv
# Imports the cosine_similarity function from scikit-learn
from sklearn.metrics.pairwise import cosine_similarity
# Imports the NumPy library,
import numpy as np
# Loads environment variables, likely including your OPENAI_API_KEY.
load_dotenv()
# Initializes an OpenAIEmbeddings object.
embedding = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=300)# Specifies the OpenAI embedding model to use.Sets the desired dimensionality of the output embeddings.
#  Defines a list of strings, where each string is a document. In this case, they are descriptions of famous Indian cricketers.
documents = [
    "Virat Kohli is an Indian cricketer known for his aggressive batting and leadership.",
    "MS Dhoni is a former Indian captain famous for his calm demeanor and finishing skills.",
    "Sachin Tendulkar, also known as the 'God of Cricket', holds many batting records.",
    "Rohit Sharma is known for his elegant batting and record-breaking double centuries.",
    "Jasprit Bumrah is an Indian fast bowler known for his unorthodox action and yorkers."
]
#  query string you want to compare against the documents.
query = 'tell me about bumrah'
# Generates embedding vectors for each document in the documents list using the initialized OpenAI embedding model.
doc_embeddings = embedding.embed_documents(documents)
# Generates an embedding vector for the query string.
query_embedding = embedding.embed_query(query)
# Calculates the cosine similarity between the query_embedding and each of the doc_embedding
scores = cosine_similarity([query_embedding], doc_embeddings)[0]
# enumerate(scores): Creates pairs of (index, score) for the list of scores.
# sorted(..., key=lambda x:x[1]): Sorts the list of tuples based on the similarity score (the second element of each tuple).
# [-1]: Selects the last element of the sorted list, which is the tuple with the highest similarity score.

index, score = sorted(list(enumerate(scores)),key=lambda x:x[1])[-1]

print(query)
print(documents[index])
print("similarity score is:", score)
```
